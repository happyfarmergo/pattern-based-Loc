```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from shapely.geometry import Point, LineString
```

```python
dataset = 'jiading'
datatype = '2g'
data_path = './data/%s_%s/' % (dataset, datatype)
disp_path = './display/%s_%s/' % (dataset, datatype)
cnn_path = data_path + 'result/cnn.csv'
df = pd.read_csv(cnn_path)
```

```python
df
```

```python
plt.figure(figsize=(5, 5))
plt.plot(df['Iteration'], [0.75] * len(df), 'r-')
plt.plot(df['Iteration'], df['Accuracy'], 'b--')
plt.axis([0, 950, 0, 1.0])
plt.legend(['RF Classifier', 'CNN Classifier'], fontsize=16)
plt.xlabel('Number of Iterations')
plt.ylabel('Accuracy')
plt.title('RF vs CNN-2')
plt.savefig(disp_path + 'result/exp.png')
```
