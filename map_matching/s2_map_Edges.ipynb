```python
import pandas as pd
df=pd.read_csv('../data/jiading_map/Nodes.csv')
df
```

```python
def get_node_lat_lon(node_id):

    node=df[df.origin_id==node_id]
    return node.iloc[0,1],node.iloc[0,2]

# print get_node_lat_lon(4727003825)


def cal_distance(nodeA,nodeB):
#     nodeA=eval(nodeA)
#     nodeB=eval(nodeB)
    from geopy.distance import vincenty
    latA,lonA=get_node_lat_lon(nodeA)
    latB,lonB=get_node_lat_lon(nodeB)
    a=(latA,lonA)
    b=(latB,lonB)

    distance=vincenty(a, b).m
    return distance
    
# print cal_distance(4727003825,4727003824)

def distance(strlist):
    list=[eval(node_id) for node_id in eval(strlist)]
    
    dis=0
    for i in range(len(list)-1):
        dis=dis+cal_distance(list[i],list[i+1])
    return dis
```

```python
ddf=pd.read_csv('../data/jiading_map/highway.csv')[['id','nodes']]
# print ddf
ddf['from']=ddf['nodes'].apply(lambda x:eval(x)[0])
ddf['to']=ddf['nodes'].apply(lambda x:eval(x)[-1])

# from __future__ import division  
# ddf['distance']=ddf['nodes'].apply(lambda x: distance(x))
# ddf['speed']=(ddf['maxspeed']+ddf['minspeed'])/2
ddf['cost']=ddf['nodes'].apply(lambda x:distance(x))
ddf
```

```python
ddff=ddf[['id','from','to','cost','nodes']] 
ddff.to_csv('../data/jiading_map/Edges.csv',index=False)
ddff.head()
```
