```python
import pandas as pd
highway_df=pd.read_csv('../data/jiading_map/highway.csv')
edge_df=pd.read_csv('../data/jiading_map/Edges.csv')
print len(highway_df), len(edge_df)
result=pd.merge(highway_df,edge_df,on=['id','nodes'])
# alldata_df['length']=alldata_df['distance']*1.609344*1000
result
```

```python
import pandas as pd
f=pd.read_csv('../data/jiading_map/Nodes.csv')

def get_node_lat_lon(node_id):
#     if (node_id.contains("'")):
#         node_id=node_id[1:-1]
    node=f[f.origin_id==node_id]
    return str(node.iloc[0,1])+'^'+str(node.iloc[0,2])

# get_node_lat_lon(4704039842)

def format(strlist):
    list=[eval(node_id) for node_id in eval(strlist)]
    
    str=''
    for i in range(len(list)):
        str=str+'^'+get_node_lat_lon(list[i])
    return str

new['strnodes']=new['nodes'].apply(lambda x: format(x))
new
```

```python
new[['id','category','distance','nodes','strnodes']].to_csv('../data/jiading_map/WA_EdgeGeometry.csv',index=False)
```
